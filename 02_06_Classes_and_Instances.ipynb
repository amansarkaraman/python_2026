## Classes

```python
class Dog:
    def __init__(self, name):
        self.name = name
        self.legs = 4
    
    def speak(self):
        print(self.name + ' says: Bark!')
        
```

```python
my_dog = Dog('Rover')
another_dog = Dog('Fluffy')
```

```python
my_dog.speak()
```

```python
another_dog.speak()
```
